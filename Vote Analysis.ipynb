# _Load and Inspect the Dataset._

```python
import pandas as pd
data = pd.read_csv("Dataset .csv")

data.head()
```

# _Analyze Correlation between Number of Votes and Rating._

```python
data['Votes'] = pd.to_numeric(data['Votes'], errors='coerce')

highest_votes = data.loc[data['Votes'].idxmax()]

lowest_votes = data.loc[data['Votes'].idxmin()]

highest_votes_info = highest_votes[['Restaurant Name', 'Votes']]
lowest_votes_info = lowest_votes[['Restaurant Name', 'Votes']]

highest_votes_info, lowest_votes_info
```

# _Analyze Correlation between Number of Votes and Rating._

```python
import matplotlib.pyplot as plt
import seaborn as sns

cleaned_data = data.dropna(subset=['Votes', 'Aggregate rating'])

correlation = cleaned_data['Votes'].corr(cleaned_data['Aggregate rating'])

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Votes', y='Aggregate rating', data=cleaned_data)
plt.title(f'Correlation between Votes and Aggregate Rating: {correlation:.2f}')
plt.xlabel('Votes')
plt.ylabel('Aggregate Rating')
plt.grid(True)
plt.show()

correlation
```

_The correlation between the number of votes and the aggregate rating is approximately 0.31. This indicates a positive but moderate correlation, meaning that generally, restaurants with more votes tend to have higher ratings, but this relationship is not very strong._
